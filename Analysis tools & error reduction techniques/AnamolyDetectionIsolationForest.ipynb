## Detecting outliers in dataset using Isolation Forest algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# creating 1000 random data points(with 2 features)
data_points = np.random.randn(1000,2)
plt.scatter(data_points[:,0], data_points[:,1], label='data points')
plt.xlabel('feature1 value')
plt.ylabel('feature2 value')
plt.show()
```

```python
# creating object of classifier to detect 3 % data points as anamolies
isolation_forest = IsolationForest(contamination=.03)
isolation_forest.fit(data_points)
```

```python
# 1 for inlier data points & -1 for outlier data points
prediction = isolation_forest.predict(data_points)
```

```python
inlier_data_points = np.nonzero(prediction == 1)
outlier_data_points = np.nonzero(prediction == -1)
```

```python
plt.scatter(data_points[inlier_data_points, 0], data_points[inlier_data_points,1], label='inliers')
plt.scatter(data_points[outlier_data_points, 0], data_points[outlier_data_points,1], label='outliers')
plt.xlabel('feature1 value')
plt.ylabel('feature2 value')
plt.legend()
plt.show()
```
